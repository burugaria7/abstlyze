```python
import numpy as np
import umap
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 3次元データのサンプルを生成
np.random.seed(42)
data = np.random.rand(100, 3)

# UMAPを使用して3次元データを2次元に埋め込む
reducer = umap.UMAP(n_neighbors=10, min_dist=0.1, n_components=2)
embedding = reducer.fit_transform(data)

# 3次元データのプロットと保存
fig = plt.figure(figsize=(6, 6))  # サイズを512x512ピクセルに設定
ax3d = fig.add_subplot(121, projection='3d')
ax3d.scatter(data[:, 0], data[:, 1], data[:, 2])
ax3d.set_title('3d_plot')
plt.savefig('3d_plot.png', dpi=100)  # 3DプロットをPNG形式で保存

# 2次元データのプロットと保存
fig = plt.figure(figsize=(6, 6))  # サイズを512x512ピクセルに設定
ax2d = fig.add_subplot(111)
ax2d.scatter(embedding[:, 0], embedding[:, 1])
ax2d.set_title('umap_plot')
plt.savefig('umap_plot.png', dpi=100)  # UMAPプロットをPNG形式で保存

plt.show()
```
